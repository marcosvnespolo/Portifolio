## Overview do Projeto


*   O objetivo do projeto é a utilização de algorítimos de ML para classificação de vinhos, identificando quais características tornam um vinho tinto ou branco;
*   A base de dadas utilizada pode ser encontrada em: https://www.kaggle.com/datasets/dell4010/wine-dataset.




```python
import pandas as pd
dataset = pd.read_csv('/content/drive/MyDrive/Datasets/Avaliação de Vinhos/wine_dataset.csv')
display(dataset)
```

Para proceder com as análises, é necessário que a coluna 'Style' seja composta apenas por valores numéricos

```python
dataset['style'] = dataset['style'].replace('red', 0)
dataset['style'] = dataset['style'].replace('white', 1)
```

```python
display(dataset)
```

Podemos separar as variáveis em preditora e variável Avaliação

```python
y = dataset['style']
x = dataset.drop('style', axis =1)
display(x, y)
```

Separando os conjustos de dados de treino e de teste

```python
from sklearn.model_selection import train_test_split
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.3)
```

Criando o modelo de ML e imprimindo o resultado

```python
from sklearn.ensemble import ExtraTreesClassifier
modelo = ExtraTreesClassifier()
modelo.fit(x_treino, y_treino)

resultado = modelo.score(x_teste, y_teste)
print('Acurácia: ', resultado)
```



*   A acurácia compara os resultados reais com os obtidos pelo mdelo;
*   Nesete caso, o modelo se demonstrou bastante satisfatório, acertando a previsão em praticamente todas as tentativas.
